```python
# Definir la lista de palabras explícitas
explícita = ['data', 'software', 'development', 'developer', 'it', 'technical', 'software', 'junior', 'senior']
```

```python
import re

def filter_by_keyword(df, keywords):
    # Crear una expresión regular para buscar coincidencias de palabras individuales
    pattern = r'\b(?:{})\b'.format('|'.join(keywords))
    # Filtrar el DataFrame utilizando la expresión regular
    filtered_df = df[df['jobTitle'].str.contains(pattern, case=False)]
    return filtered_df
```

```python
df_filtred= filter_by_keyword(df,explícita)
```

```python
df_filtred
```

```python
df_filtred['jobTitle'].value_counts().head(50)
```

```python
len(df_filtred['jobTitle'].unique())
```

```python
def get_category(row):
    title = row['jobTitle'].lower()
    if 'software engineer' in title:
        return 'software engineer'
    elif 'data engineer' in title:
        return 'data engineer'
    elif 'data scientist' in title:
        return 'data scientist'
    elif 'data analyst' in title:
        return 'data analyst'
    else:
        return 'IT'

df_filtred['category'] = df_filtred.apply(get_category, axis=1)
```

```python
df_filtred['category'].value_counts()
```
