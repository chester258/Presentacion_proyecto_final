Usando un csv aparte podemos llevar a cabo ciertos metodos para conseguir el codigo pais de cada ciudad, a partir de la latitud y longitud

```python
extra = pd.read_csv(r'Dataset\worldcities.csv')
```

```python
extra
```

```python
code = extra[['city','iso2','lat','lng']]

#Sepramaos las columnas que nos interesan
```

```python
code
```

```python
Location.reset_index(inplace=True)
```

```python
Location
```

```python

location_fin = Location.merge(code,how='left', on="city")

#Combinamos los 2 dataframe a partir de la columna city, aunque esto va a traer ciertas columnas 
#que se duplican ya que hay varias ciudades, hay varias ciudades con el mismo nombre por todo el mundo
#Para corregirlo vamos a basarnos en las columnas latitud y longitud 
```

```python
location_fin
```

```python
location_fin['Lat_value'] = location_fin['map.lat'] - location_fin['lat']
```

```python
location_fin['lng_value'] = location_fin['map.lng'] - location_fin['lng']
```

```python
location_fin
```

```python
location_fin = location_fin[~((location_fin['Lat_value'] > 1) | (location_fin['Lat_value'] < -1) | (location_fin['lng_value'] > 1) | (location_fin['lng_value'] < -1))]

#No usamos las columnas de latitud y longitud para unir los dataframes porque no 
#son exactamente iguales los valores que manejan, por lo tanto llevamos a cabo este metodo.
```

```python
location_fin
```

```python
location_fin.dropna(subset=['iso2'],inplace=True)
```

```python
location_fin
```

```python
location_fin_c = location_fin[['index','map.employerName', 'map.lat', 'map.lng', 'city', 'iso2']]
```

```python
location_fin_c.rename(columns={'iso2':'Code'},inplace=True)
```

```python
location_fin_c
```

```python
coutry_n = pd.read_csv(r'Dataset\country_names_2_digit_codes.csv')
```

```python
coutry_n
```

```python
#Vemos si contiene datos nulos
coutry_n.isna().sum()
```

```python
#Buscamos el error
coutry_n[coutry_n['Code'].isna()]
```

```python
#Corregimos el valor
coutry_n.fillna("NA",inplace=True)
```

```python
location_fin_c = location_fin_c.merge(coutry_n,how='left', on="Code")
```

```python
location_fin_c
#Asi obtuvimos los paise 
```

```python
location_fin_c.to_csv("location_fin_clean.csv",index=False)
```
